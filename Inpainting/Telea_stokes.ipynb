```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('../Images/IR_00002.jpg', cv2.IMREAD_COLOR)


mask = np.zeros(image.shape[:2], dtype=np.uint8) 
mask[50:200, 50:200] = 255  


inpaint_telea = cv2.inpaint(image, mask, 3, cv2.INPAINT_TELEA)

# Inpainting with Navier-Stokes method
inpaint_ns = cv2.inpaint(image, mask, 3, cv2.INPAINT_NS)

# Display the images
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(cv2.cvtColor(inpaint_telea, cv2.COLOR_BGR2RGB))
plt.title('Inpainting with Telea\'s method')
plt.axis('off')

plt.subplot(133)
plt.imshow(cv2.cvtColor(inpaint_ns, cv2.COLOR_BGR2RGB))
plt.title('Inpainting with Navier-Stokes method')
plt.axis('off')
plt.show()
```
